```python
from huggingface_hub import hf_hub_download
from ultralytics import YOLO

model_path = hf_hub_download(repo_id="AdamCodd/YOLOv11n-face-detection", filename="model.pt")
model = YOLO(model_path)

```

```python
results = model.predict("C:\\Users\\user\\OneDrive\\Documents\\Discord-Bot\\pexels-photo-2379005.jpeg", save=True) # saves the result in runs/detect/predict
```

```python
# Print a summary and extract per-detection confidences (and labels if available)
if not results:
    print("No results returned.")
else:
    for i, r in enumerate(results):
        # r.boxes.conf should hold the confidence for each detection
        confs = []
        try:
            # Most ultralytics versions store a tensor here
            confs = r.boxes.conf.cpu().numpy().tolist()
        except Exception:
            try:
                confs = [float(x) for x in r.boxes.conf]
            except Exception:
                # Last-resort fallback: coerce to list
                confs = list(r.boxes.conf)
        print(f"Image {i} confidences: {confs}")
        # If class labels are available, print them alongside confidences
        if hasattr(r.boxes, 'cls'):
            try:
                labels = r.boxes.cls.cpu().numpy().astype(int).tolist()
            except Exception:
                labels = list(r.boxes.cls)
            print(f"Image {i} labels: {labels}")
```
