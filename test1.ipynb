```python
from huggingface_hub import hf_hub_download
from ultralytics import YOLO

model_path = hf_hub_download(repo_id="AdamCodd/YOLOv11n-face-detection", filename="model.pt")
model = YOLO(model_path)
```

```python
results = model.predict("C:\\Users\\user\\OneDrive\\Documents\\Discord-Bot\\pexels-photo-2379005.jpeg", save=True) # saves the result in runs/detect/predict
```

```python
def get_confidence(results):
    for i,r in enumerate(results):
        print(f"--- Results for image {i} ---")
        for box in r.boxes:
            conf = box.conf.item()  # get confidence
            cls = box.cls.item()    # get class
            print(f"Class: {cls}, Confidence: {conf:.4f}")

if ( get_confidence(results)):
    print("delete the image.")
```
